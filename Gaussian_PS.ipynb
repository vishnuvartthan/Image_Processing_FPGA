```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
og_image = cv2.imread('bellatrix.jpg', cv2.IMREAD_GRAYSCALE)

s = 0.6

image = cv2.resize(og_image, (0, 0), fx=s, fy=s, interpolation=cv2.INTER_AREA)

height, width = image.shape

r = 1
kernel_size = (r*2+1, r*2+1)

sigma = 0.5
```

```python

start_time = time.time()

gaussian_blurred_image = cv2.GaussianBlur(image, kernel_size, sigma)


end_time = time.time()
elapsed_time = (end_time - start_time) * 1000


print(f"Image size       : {width} x {height} pixels")
print(f"Kernel size      : {kernel_size[0]} x {kernel_size[1]}")
print(f"Gaussian blur (σ={sigma}) execution time: {elapsed_time:.3f} ms")
```

```python
plt.figure(figsize=(20, 20))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gaussian_blurred_image, cmap='gray')
plt.title(f"Gaussian Blur ({kernel_size[0]}x{kernel_size[1]}, σ={sigma})")
plt.axis('off')
plt.show()
```
