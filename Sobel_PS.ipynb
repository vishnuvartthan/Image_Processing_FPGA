```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
image = cv2.imread('bellatrix.jpg', cv2.IMREAD_GRAYSCALE)

s = 0.5

scaled_image = cv2.resize(image, (0, 0), fx=s, fy=s, interpolation=cv2.INTER_AREA)

height, width = scaled_image.shape

kernel = np.array([[1.0,0.0,-1.0],[2.0,0.0,-2.0],[1.0,0.0,-1.0]],np.float32)

```

```python
start_time = time.time()

outImage = cv2.filter2D(scaled_image, -1, kernel)


end_time = time.time()
elapsed_time = (end_time - start_time) * 1000 
```

```python
print(f"Image size       : {width} x {height} pixels")
print(f"Sobel edge detection execution time: {elapsed_time:.3f} ms")
```

```python
plt.figure(figsize=(25, 25))
plt.subplot(1, 2, 1)
plt.imshow(scaled_image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(outImage, cmap='gray')
plt.title(f"Sobel Edge Detection")
plt.axis('off')
plt.show()
```
